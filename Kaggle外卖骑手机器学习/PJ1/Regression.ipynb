```python
import pandas as pd
import numpy as np
import random
from matplotlib import pyplot as plt
import seaborn as sns


%matplotlib inline
sns.set()
random.seed(666)
```

```python
df_train = pd.read_csv('dataframe_train.csv')
df_test = pd.read_csv('dataframe_test.csv')
```

# Regression

# Question(a)

```python
# remove outliers
df_train = df_train.drop(df_train[df_train.grid_distance>10000].index)

# down-sample randomly
df = df_train.sample(n=10000,random_state=666,axis=0)

# drop unreasonable variables and variables we can't use in test set
df = df.drop(columns = ['courier_id', 'wave_index', 'tracking_id','date','group',
                        'id','shop_id','aoi_id','source_tracking_id','action_type'])

# convert into numerical data
df = pd.get_dummies(df)

# Standardize
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler().fit(df)
df_standard = scaler.transform(df)
df_standard = pd.DataFrame(data=df_standard,columns=df.columns)

df['grid_distance'] = df_standard['grid_distance'].values
df['urgency'] = df_standard['urgency'].values


fig = plt.figure(figsize=(10,10))
sns.heatmap(df.corr())
```

```python
df.corr()['expected_use_time']
```

We choose the following variables as independent variables:

'speed', 'grid_distance','urgency','weather_grade_Normal Weather', 'weather_grade_Very Bad Weather', 'source_type_ASSIGN','source_type_PICKUP'

# Question(b)

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import train_test_split as tr_te_split

X = df[['speed', 'grid_distance','urgency','weather_grade_Normal Weather',
        'weather_grade_Very Bad Weather', 'source_type_ASSIGN','source_type_PICKUP']]
y = df['expected_use_time'].ravel()
X_train, X_test, y_train, y_test = tr_te_split(X, y, test_size=0.3, random_state=666)

reg = LinearRegression()
param_grid = [{'fit_intercept':[0,1],
              'positive':[0,1]}]
grid = GridSearchCV(reg, param_grid,cv=5)
grid.fit(X_train,y_train)
print("The best parameter is",grid.best_params_) 
print('\n') 
print("The corresponding R-squared is %0.4f." % grid.best_score_) 
```

```python
from sklearn.metrics import mean_absolute_error
reg = LinearRegression(fit_intercept=1,positive=0).fit(X_train, y_train)
y_pred = reg.predict(X_test)
print('MAE of baseline model is:',mean_absolute_error(y_test, y_pred))
```

# Question(c)

K-NN

```python
from sklearn.neighbors import KNeighborsRegressor

# We use the mean and standard deviation of the training set to standardize both the training and testing sets.
scaler = StandardScaler().fit(X_train)
X_train_st = scaler.transform(X_train)
X_test_st = scaler.transform(X_test)

knn_reg = KNeighborsRegressor()
param_grid = [{'n_neighbors':[5,10,20,30,50,100],
              'p':[1,2],
              'weights':['uniform', 'distance']}]
grid = GridSearchCV(knn_reg, param_grid,cv=5)
grid.fit(X_train_st,y_train)
print("The best parameter is",grid.best_params_) 
print('\n') 
print("The corresponding R-squared is %0.4f." % grid.best_score_) 
```

```python
knn_reg = KNeighborsRegressor(n_neighbors=50,p=1,weights='distance').fit(X_train_st, y_train)
y_pred = knn_reg.predict(X_test_st)
print('MAE of baseline model is:',mean_absolute_error(y_test, y_pred))
```

Random forest

```python
from sklearn.ensemble import RandomForestRegressor
rfr = RandomForestRegressor()
param_grid = [{
    'max_depth':[5,15],
    'min_samples_split':[2,8,16],
    'n_estimators':[100,500]
}]
grid = GridSearchCV(rfr, param_grid, cv=5)
grid.fit(X_train, y_train)
print("The best parameter is",grid.best_params_) 
print('\n') 
print("The corresponding R-squared is %0.4f." % grid.best_score_) 
```

```python
rfr = RandomForestRegressor(max_depth=5,min_samples_split=16,n_estimators=500).fit(X_train,y_train)
y_pred = rfr.predict(X_test)
print('MAE of baseline model is:',mean_absolute_error(y_test, y_pred))
```

XGBT

```python
import xgboost as xgb
xgb_reg = xgb.XGBRegressor()
param_grid = [{
    'objective':['reg:squarederror'],'use_label_encoder':[False],
    'learning_rate':[0.1,0.2], 'gamma':[0.001,0.002],'colsample_bynode':[0.8],
    'max_depth':[5,15],  'n_estimators':[100,500]
}]
grid = GridSearchCV(xgb_reg, param_grid, cv=5)
grid.fit(X_train, y_train)
print("The best parameter is",grid.best_params_) 
print('\n') 
print("The corresponding R-squared is %0.4f." % grid.best_score_) 
```

```python
xgb_reg = xgb.XGBRegressor(objective='reg:squarederror',use_label_encoder=False,learning_rate=0.1,
                           gamma=0.001,colsample_bynode=0.8,max_depth=5,n_estimators=100).fit(X_train,y_train)
y_pred = xgb_reg.predict(X_test)
print('MAE of baseline model is:',mean_absolute_error(y_test, y_pred))
```

# Question(d)

Polynomial Features

```python
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import Ridge

poly_pipe = make_pipeline(PolynomialFeatures(degree=1), RandomForestRegressor(max_depth=5,min_samples_split=16,n_estimators=500))
poly_pipe.fit(X_train, y_train)

print("The out-of-sample R-squared with polynomial features and ridge regression is %0.4f."%poly_pipe.score(X_test, y_test))
```
